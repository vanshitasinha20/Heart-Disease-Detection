##Import the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##Data Collection and Processing

```python
heart_data= pd.read_csv('/content/heart.csv')
```

1--> Defective Heart

## Spilittng the Features and the Target

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

##Spiltting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
#stratify=Y is used to evenly distribute the dataset into test and train and random state=2(decides the randomness of the values)
```

```python
print(X.shape, X_train.shape, X_test.shape )
```

## Model training

##Logistic Regression


```python
model= LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)#finds relationship b/w the features and the target
```

## Model Evaluation


Accuracy Checking


```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy of the training data is:-', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy of the test data is:-', testing_data_accuracy)
```

##Building a Predictive **System**

```python
input_data=(71,0,0,112,149,0,1,125,0,1.6,1,0,2)

#change the input data to numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for only one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1 )
prediction=model.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):#prediction[0] means the 0th index value i.e 0
  print('The Person does not have a Heart Disease')
else:
  print('The Person has a Heart Disease')
```
